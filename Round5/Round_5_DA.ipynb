```python
def get_time_trades(df, name):
    buy = df[df['buyer'] == name]
    sell = df[df['seller'] == name]
    selling = list(sell.index)
    buying = list(buy.index)
    for i in buying:
        plt.axvline(i,color = 'green', markersize = 0.0000001)
    for i in selling:
        plt.axvline(i,color = 'red', markersize = 0.0000001)
```

```python
for i in set(star.seller):
    print(i)
    get_info(star, i)
    get_time_trades(star, i)
    plt.show()
```
