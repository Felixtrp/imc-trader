```python
def get_info(df, name):
    buy = df[df.buyer == name].price
    sell = df[df.seller == name].price
    plt.plot(buy, color = 'blue')
    plt.plot(sell, color = 'red')
```

```python
def get_index_trades(df, name):
    buy = df[df['buyer'] == name]
    sell = df[df['seller'] == name]
    selling = list(sell.index)
    buying = list(buy.index)
    for i in buying:
        plt.axvline(i,color = 'green', markersize = 0.0000001)
    for i in selling:
        plt.axvline(i,color = 'red', markersize = 0.0000001)

def get_time_trades(df, name):
    buy = df[df['buyer'] == name]
    sell = df[df['seller'] == name]
    selling = list(sell.timestamp)
    buying = list(buy.timestamp)
    for i in buying:
        plt.axvline(i,color = 'green', markersize = 0.0000001)
    for i in selling:
        plt.axvline(i,color = 'red', markersize = 0.0000001)

def get_time_buy(df, name):
    buy = df[df['buyer'] == name]
    sell = df[df['seller'] == name]
    selling = list(sell.timestamp)
    buying = list(buy.timestamp)
    for i in buying:
        plt.axvline(i,color = 'green', markersize = 0.0000001)

def get_time_sell(df, name):
    buy = df[df['buyer'] == name]
    sell = df[df['seller'] == name]
    selling = list(sell.timestamp)
    buying = list(buy.timestamp)
    for i in selling:
        plt.axvline(i,color = 'red', markersize = 0.0000001)
```
