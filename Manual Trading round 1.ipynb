```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
area = lambda x: (1/10000) * (x - 900)**2
def profit(a, b):
    low = area(a)
    high = area(b)
    return (1000 - a) * low + (1000 - b) * (high - low)
best = [0, (0, 0)]
for i in range(900, 1001):
    for j in range(i, 1001):
        curr = profit(i, j)
        if best[0] < curr:
            best = curr, (i, j)
print(best)
```

```python
def inverse_cdf(u):
    return np.sqrt((u) * 10000) + 900

uniform_samples = np.random.rand(10000)
samples = inverse_cdf(uniform_samples)

plt.hist(samples, bins=10, color='grey', edgecolor='black')
plt.xlabel('Prices')
plt.ylabel('Frequency')
plt.title('Histogram of Prices')
plt.show()
```

```python
n = 5
profits = [0] * n
lst = [i - (n // 2) for i in range(n)]
uniform_samples = np.random.rand(3000)
samples = inverse_cdf(uniform_samples)
for sample in samples:
    for i, diff in enumerate(lst):
        if sample <= 952 + diff:
            profits[i] += 1000 - (952 + diff)
        if (978 + diff) >= sample > (952 + diff):
            profits[i] += 1000 - (978 + diff)
plt.axvline(x=0)
plt.plot(lst, profits)
```
